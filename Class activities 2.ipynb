```python
#Standardizing header names
# Rename a single column
df.rename(columns={'GENDER': 'Gender'}, inplace=True)
```

```python
df.head()
```

```python
#Replace all occurences of words starting with f, like "female" with simply the letter F
df['Gender'] = df['Gender'].str.replace(r'^f', 'F', regex=True)
```

```python
#Replace all occurences of words starting with m, like "male" with simply the letter M
df['Gender'] = df['Gender'].str.replace(r'^m', 'M', regex=True)
```

```python
# Replace values that are neither "M" nor "F" with "NA"
df.loc[~df['Gender'].isin(['M', 'F']), 'Gender'] = 'NA'
```

```python
df.head()
```

```python
df.describe()
```

```python
# Working with data types â€“ Check the data types of all the columns and fix the incorrect ones (for ex. customer lifetime value and number of complaints )

# Remove non-numeric characters from the column
df['Customer Lifetime Value'] = df['Customer Lifetime Value'].str.replace('%', '')

# Identify non-numeric values in the column
non_numeric_values = pd.to_numeric(df['Customer Lifetime Value'], errors='coerce').isna()

#Replace empty rows and non numeric values with a default value
default_value = 0
df['Customer Lifetime Value'].fillna(default_value, inplace=True)
# Replace non-numeric values with default value
df.loc[non_numeric_values, 'Customer Lifetime Value'] = default_value

#Convert the column to float
df['Customer Lifetime Value'] = df['Customer Lifetime Value'].astype(float)
```

```python
df.head()
```

```python
# Remove non-numeric characters from the column
df['Number of Open Complaints'] = df['Number of Open Complaints'].str.replace('/', '')
```

```python
df.head()
```

```python
#Removing duplicates
# Remove duplicates based on 'Customer'
df.drop_duplicates(subset=['Customer'], inplace=True)
```

```python
#Write a function to replace column "State" to different zones. California as West Region, Oregon as North West, and Washington as East, and Arizona and Nevada as Central
def replace_state_zones(df):
    # Create a dictionary to map states to zones
    state_zones = {
        'California': 'West Region',
        'Oregon': 'Northwest',
        'Washington': 'East',
        'Arizona': 'Central',
        'Nevada': 'Central'
    }
    
    # Replace the values in the "State" column with the corresponding zones
    df['ST'].replace(state_zones, inplace=True)
    
    # No need to return the DataFrame as it has been modified in-place
    
# Call the function with the DataFrame to update the "State" column
replace_state_zones(df)
```
